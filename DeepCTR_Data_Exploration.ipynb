```python
from datetime import datetime
import numpy as np
import os
import pandas as pd

path = '/content/drive/My Drive/DAAN 570'
os.chdir(path)
df = pd.read_csv("df_final.csv")
```

```python
# prompt: drop pageview	detail	event	click	add	purchase remove

df.drop(columns=['pageview', 'detail', 'event', 'click', 'add', 'purchase', 'remove'], axis=1, inplace=True)
df.head()
```

```python
import pandas as pd

# Calculate descriptive statistics
descriptive_stats = df.describe()

# Calculate the range for each numerical column
range_df = df.select_dtypes(include=[np.number]).apply(lambda x: x.max() - x.min())

# Add the range information to the descriptive statistics
descriptive_stats.loc['range'] = range_df

print(descriptive_stats)
```

```python
# Check for missing values
missing_values = df.isnull().sum()

# Percentage of missing values
missing_percentage = (df.isnull().sum() / len(df)) * 100

# Display missing values and their percentages
missing_info = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})
missing_info = missing_info[missing_info['Missing Values'] > 0]  # Filter columns with missing values
print(missing_info)

# Check for any duplicate rows
duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")
```

```python
# Assessing imbalance in categorical columns
for column in df.select_dtypes(include=['object']).columns:
    print(f"\nDistribution in column {column}:")
    print(df[column].value_counts(normalize=True))  # Provides percentages of each category
```

```python
import numpy as np
# Adjusting the approach to explicitly exclude non-numeric columns from the quantile calculations
numeric_columns = df.select_dtypes(include=[np.number])

# Recalculating the IQR method components for numeric columns
Q1 = numeric_columns.quantile(0.25)
Q3 = numeric_columns.quantile(0.75)
IQR = Q3 - Q1

# Recalculating the boundaries for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Reidentifying outliers
outliers_corrected = ((numeric_columns < lower_bound) | (numeric_columns > upper_bound)).sum()

# Redisplaying the number of outliers in each numeric column
outliers_corrected[outliers_corrected > 0]  # Filtering columns with outliers


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Setting up the figure size and layout
plt.figure(figsize=(18, 12))
plt.subplots_adjust(hspace=0.5)

# Plotting histograms for numerical columns to visualize distributions and outliers
for index, column in enumerate(numeric_columns.columns, start=1):
    plt.subplot(3, 3, index)
    sns.histplot(numeric_columns[column], kde=True)
    plt.title(column)

plt.show()
```

```python
# Calculate the correlation matrix
corr_matrix = numeric_columns.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```
